```python
import gymnasium as gym
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.distributions import Categorical
import numpy as np
from tqdm import trange
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import random
import time

# plt.ion()

mps_device = torch.device("cpu")
n_hidden = 512

max_ret = 0
max_rets, ep_rets, run_avg = [], [], []
window = 100
# lr = 1e-3
episodes = 500
discount=1
batch_size = 512
    # render = True


class DQN(nn.Module):
    def __init__(self, fan_in, fan_out):
        self.fan_in = fan_in
        self.fan_out = fan_out
        super().__init__()
        self.block = nn.Sequential(*[
            nn.Linear(fan_in, n_hidden, bias = True),
            nn.ReLU(),
            nn.Linear(n_hidden, n_hidden, bias = True),
            nn.ReLU(),
            nn.Linear(n_hidden, n_hidden, bias = True),
            nn.ReLU(),
            nn.Linear(n_hidden, fan_out, bias = True),
        ])
       
        self.optim = torch.optim.AdamW(self.parameters())
    def forward(self, x):
        if self.fan_out == 1:
            return self.block(x)
        else:   
            logits = self.block(x)
            return Categorical(logits=logits)


class Agent:
    def __init__(self):
        super().__init__()
        self.policy = DQN(8,4)
        self.value = DQN(8,1)
        self.value.to(mps_device)
        self.policy.to(mps_device)
        self.history = []
        self.max_rew = 0

    def act(self, obs):
        return self.policy(obs).sample().item(), self.value(obs).item()

    def record(self, episode):
        self.history.append(episode)

    def update(self, obs, act, wei):

        # val_loss = torch.mean(val_loss_sum)
        # val_loss.requires_grad = True
        # self.value.optim.zero_grad()
        # val_loss.backward()
        # self.value.optim.step()

        # samp = torch.randint(len(obs), (batch_size,))
        
        logp = self.policy(obs).log_prob(act)

        # wei = torch.tensor(np.array([r - v for r, v in zip(rtg, val)]))
        # print(rtg.shape, val.shape, len(wei), logp.shape)
        
        # print(logp.shape, wei.shape, act.shape, obs.shape)
        loss = (-logp * wei).mean()
        # loss.requires_grad = True
        self.policy.optim.zero_grad()
        loss.backward()
        self.policy.optim.step()
        # self.history.clear()
        


lander = Agent()



def plot_results(x, y):

    clear_output(wait=True)
    max = [np.max(y)] * len(x)
    cnt = len(x)

    if cnt > window:
        run_avg.append(np.mean(x[-window:]))
         
    plt.figure(figsize=(16, 4))
    plt.plot(x, label = 'returns')
    plt.plot(range(window, len(x)), run_avg, label = 'avg returns')
    plt.plot(y, label = 'max')
    
    # print(f'ep_ret {x[-1]:6.2f} | max {max[-1]:6.2f}')
    
    # plt.text(-10, max[-1], f'{max[-1]:.2f}', color='orange', fontsize=12, ha='center', va='center')
    # plt.text(-10, x[-1], f'{x[-1]:.2f}', color='blue', fontsize=12, ha='center', va='center')

    plt.xlabel('episodes')
    plt.ylabel('rewards')
    plt.show()
    # plt.draw()
    # plt.pause(1)
    # plt.clf()
    # time.sleep(1)
    

def reward_to_go(rews):
    n = len(rews)
    rtgs = np.zeros_like(rews)
    for i in reversed(range(n)):
        rtgs[i] = rews[i] + discount * (rtgs[i+1] if i+1 < n else 0)
    return list(rtgs)

def train(epoch):
  
    global max_ret
    global ep_rets
    
    batch_obs = []
    batch_act = []
    batch_wei = []
    val_loss_sum = 0
    # rollouts
    for ep in range(episodes):
        ep_rews = []
        ep_vals = []
        env = gym.make("LunarLander-v2", render_mode = ("human" if ep==0 else None)) 
        observation, info = env.reset()

        while True:
            batch_obs.append(observation.copy())
            
            action, value = lander.act(torch.tensor(observation, device=mps_device))
            observation, reward, terminated, truncated, info = env.step(action)
            
            batch_act.append(action)
            ep_vals.append(value)
            ep_rews.append(reward)
                

            
            if terminated or truncated:
                ep_ret, ep_len = sum(ep_rews), len(ep_rews)
                ep_rets.append(ep_ret)
                max_rets.append(max(ep_rets))

                
                plot_results(ep_rets, max_rets)
                #t.set_description(f'ep {ep + epoch*episodes} | max {max_ret:8.2f}') 
    
                rtg = torch.tensor(reward_to_go(ep_rews), dtype =torch.float32, device=mps_device)
                batch_wei += rtg
                # val_loss_sum += F.mse_loss(torch.tensor(ep_vals, device=mps_device), rtg)
                
                # print(len(batch_wei), len(batch_obs), len(batch_act))
                # batch_wei += [ep_ret] * ep_len
                # print(len(batch_wei), len(batch_obs), len(batch_act))
                ep_vals = []
                ep_rews = []

               
                
                observation, info = env.reset()
                
                break

                
    lander.update(torch.tensor(np.array(batch_obs), device=mps_device), 
                  torch.tensor(np.array(batch_act), dtype = torch.int32, device=mps_device), 
                  torch.tensor(np.array(batch_wei), dtype=torch.float32, device=mps_device),)
                    
    
    

    # env.close()
```

```python
for epoch in range(200):
    try:
        train(epoch)
    except KeyboardInterrupt:
        break

  

```
